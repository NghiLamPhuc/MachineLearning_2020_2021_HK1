```python
import numpy as np
import matplotlib.pyplot as plt

input = np.linspace(-10, 10)
# print(len(input)) #50

def sigmoid(X):return 1/(1+np.exp(-X))

output = sigmoid(input)
plt.plot(input, output, c="b")
output = sigmoid(0.5 * input)
plt.plot(input, output, c="r")
output = sigmoid(2.5 * input)
plt.plot(input, output, c="g")
```

```python
output = sigmoid(input)
plt.plot(input, output, c="b")
output = sigmoid(0.5*input + 2)
plt.plot(input, output, c="r")
output = sigmoid(2.5*input - 2)
plt.plot(input, output, c="g")
```
